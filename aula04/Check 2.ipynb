```python
#Variáveis quantitativas:
    
#V9058
#V4742
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import os
notas = pd.read_table("pes2013A.txt", encoding="windows-1252", sep=" ")
notas.columns
notas.loc[:,('UF','V4742','V8005')]
notas.REGIAO = notas.UF
notas.UF = notas.UF.astype('category')
notas.UF.cat.categories = ('RO','AC','AM','RR','PA','AP','TO','MA','PI','CE','RN','PB','PE','AL','SE','BA','MG','ES','RJ','SP','PR','SC','RS','MS','MT','GO','DF')
```

```python
notas.REGIAO[(notas.UF == 'GO') | (notas.UF == 'MT') | (notas.UF == 'MS') | (notas.UF == 'DF')]  = "Centro Oeste"
notas.REGIAO[(notas.UF == 'AC') | (notas.UF == 'AM') | (notas.UF == 'AP') | (notas.UF == 'PA') | (notas.UF == 'RR') | (notas.UF == 'RO') | (notas.UF == 'TO')]  = "Norte"
notas.REGIAO[(notas.UF == 'RS') | (notas.UF == 'PR') | (notas.UF == 'SC')]  = "Sul"
notas.REGIAO[(notas.UF == 'ES') | (notas.UF == 'MG') | (notas.UF == 'RJ') | (notas.UF == 'SP')]  = "Sudeste"
notas.REGIAO[(notas.UF == 'AL') | (notas.UF == 'BA') | (notas.UF == 'CE') | (notas.UF == 'MA') | (notas.UF == 'PI') | (notas.UF == 'PE') | (notas.UF == 'PB') | (notas.UF == 'RN') | (notas.UF == 'SE')]  = "Nordeste"
```
