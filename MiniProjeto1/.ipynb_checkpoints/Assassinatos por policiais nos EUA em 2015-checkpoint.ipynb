```python
Tabela = pd.read_csv('the-counted-2015.csv', encoding="windows-1252", sep= r'","', names = ["uid", "name", "age", "gender", "ethnicity", "month", "day", "year", "streetaddress", "city", "state", "classification", "law enforcement agency", "armed"])
tabela = Tabela.drop([0])
tabela
```

```python
Idade = tabela.age.convert_objects(convert_numeric=True)
Idade.plot.hist()
```

```python
Idade.describe()
```

```python
Idade.median()
```

```python
Idade.mode()
```

média > mediana > moda

A distribuição é assimétrica à direita porque a moda é menor que a mediana, que é menor que a média aritmética.
Nesse caso, a curva da distribuição tem assimetria positiva.

Segundo o formato do histograma obtido com os dados do DataSet estudado, é razoável pensar que as curvas da distribuição Binomial e da distribuição de Bernoulli representam o histograma obtido nesse estudo.

Porém, na distribuição de Poisson, a variância é igual a sua média, e isso não é visto nos resultados obtidos com o nosso estudo.

```python
Idade.var()
```

Além das duas distribuições teóricas acima estudadas, a distribuição Log normal também parece representar muito bem o histograma das idades obtido.
