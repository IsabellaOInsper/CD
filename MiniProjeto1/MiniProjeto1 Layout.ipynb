### Itens 4 a 6 - Resolução:

```python
Idade = Idade[(Idade > 0) & (Idade < 100)]
```

média > mediana > moda

A distribuição é assimétrica à direita porque a moda é menor que a mediana, que é menor que a média aritmética.
Nesse caso, a curva da distribuição tem assimetria positiva.

Segundo o formato do histograma obtido com os dados do DataSet estudado, é razoável pensar que as curvas da distribuição "Burr" e da distribuição "Logarítmica normal" representam o histograma obtido nesse estudo.


```python
#Tentativa de estimar os valores dos parâmetros com LogNormal
arange = np.arange(5, 88, 1)
Idade.plot.hist(normed = 1, color = "pink")
TentaLog = stats.lognorm.pdf(arange, s = 0.37, loc = -1.9, scale = 35)
plt.plot(arange, TentaLog, "-", color = "purple")
plt.show()
```

```python
#Pegar dados com o fit
x = stats.lognorm.fit(Idade)
x
```

```python
#Distribuição com Log Normal
arange = np.arange(5, 88, 1)
Log = stats.lognorm.pdf(arange, scale = x[2], loc = x[1], s = x[0])
plt.plot(arange, Log, "-", color = "red")
plt.show()
```

```python
Idade.plot.hist(normed = 1, color = "pink")
arange = np.arange(5, 88, 1)
Log = stats.lognorm.pdf(arange, scale = x[2], loc = x[1], s = x[0])
plt.plot(arange, Log, "-", color = "red")
plt.show()
```

```python
#Tentativa de estimar os valores dos parâmetros com Burr
arange = np.arange(5, 88, 1)
Idade.plot.hist(normed = 1, color = "gray")
TentaB = stats.burr.pdf(arange, 3.81, 1, loc= -0.9, scale= 31)
plt.plot(arange, TentaB, "-", color = "purple")
plt.show()
```

```python
# fit com Burr
z = stats.burr.fit(Idade)
z
```

```python
#Distribuição com Burr
#arange, 3.81, 1, loc= -0.9, scale= 31)
Burr = stats.burr.pdf(arange, z[0], z[1], loc= z[2], scale= z[3])
plt.plot(arange, Burr, "-", color = "blue")
```

```python
#Distribuição com Burr
Idade.plot.hist(normed = 1, color="gray")
Burr = stats.burr.pdf(arange, z[0], z[1], loc= z[2], scale= z[3])
plt.plot(arange, Burr, "-", color = "blue")
plt.show()
```

Com a comparação acima, acredito que os parâmetros obtidos pela função fit são os que fazem a curva se aproximar mais do histograma.

```python
#Comparação Burr
Idade.plot.hist(normed = 1, color = "gray")
TentaB = stats.burr.pdf(arange, 3.81, 1, loc= -0.9, scale= 31)
plt.plot(arange, TentaB, "-", color = "purple")
Burr = stats.burr.pdf(arange, z[0], z[1], loc= z[2], scale= z[3])
plt.plot(arange, Burr, "-", color = "blue")
plt.show()
```

Mais uma vez, com a comparação dos parâmetros escolhidos por mim e os dados pelo fit, acredito que os valores obtidos pela função fit formam uma curva que se aproxima mais da dos dados do histograma.

```python
#QQplot
X = pd.Series(sorted(Idade.dropna()))

medlogfit = Log.mean()
medlogmeu = TentaLog.mean()
medburrfit = Burr.mean()
medburrmeu = TentaB.mean()
#fra dos dados
fra = []
for i in range(len(X)):
    acc = (i - 0.5) / len(X)
    if acc < 0:
        acc = 0
    fra.append(acc)
fra = pd.Series(data=fra)
qlogfit = stats.lognorm.ppf(fra, scale = x[2], loc = x[1], s = x[0])
#qlogmeu = stats.lognorm.ppf(fra, s = 0.37, loc = -1.9, scale = 35)
qburrfit = stats.burr.ppf(fra, c = z[0], d = z[1], loc= z[2], scale= z[3])
#qburrmeu = stats.burr.ppf(fra, c = 3.81, d = 1, loc= -0.9, scale= 31)

fig = plt.figure(figsize=(12,8))
plt.title("Q-Q plot: Dados x Burr x LogNormal")
logf_plot = plt.plot(X, qlogfit, 'o', color="red", label="LogNormal com Fit")
#logm_plot  = plt.plot(X, qlogmeu, 'o', color="pink", label="LogNormal com meus parâmetros")
burrf_plot = plt.plot(X, qburrfit, 'o', color="blue", label="Burr com fit")
#burrm_plot = plt.plot(X, qburrmeu, 'o', color="purple", label="Burr com meus parâmetros")
#plt.legend(handles=[logf_plot, logm_plot])
plt.plot((min(X), max(X)), (min(X), max(X)), '-')


for i in range(len(X)):
    plt.plot((X[i], X[i]), (qlogfit[i], X[i]), '-', color='red', alpha=0.3  )
#    plt.plot((X[i], X[i]), (qlogmeu[i], X[i]), '-', color='pink', alpha=0.3  )
    plt.plot((X[i], X[i]), (qburrfit[i], X[i]), '-', color='blue', alpha=0.3  )
#    plt.plot((X[i], X[i]), (qburrmeu[i], X[i]), '-', color='purple', alpha=0.3  )
```

Essa é a diferença entre os dados e os quantis teóricos das distribuições analisadas até agora. Percebemos que as bolinhas de cor vermelha são as que menos distam da curva dos dados, ou seja a distribuição logarítimica normal é a que menos dista dos dados.

### Itens 10 e 11 - Resolução:

```python
# fra e FDA
cdf_logfit = stats.lognorm.cdf(X, scale = x[2], loc = x[1], s = x[0]) #FDA
cdf_burrfit = stats.burr.cdf(X, c = z[0], d = z[1], loc= z[2], scale= z[3])


fig = plt.figure(figsize=(20,12))

plt.title("Cumulativa: Dados x LogNormal x Burr")

dados_fra_plot, = plt.plot(X, fra, '-', color='black', label="F.R.A dos dados")
cdf_logfit_plot,   = plt.plot(X, cdf_logfit, '-', color='red', label = "Cumulativa da LogNormal")
cdf_burrfit_plot, = plt.plot(X, cdf_burrfit, '-', color='blue', label = "Cumulativa da Burr")

plt.legend(handles=[dados_fra_plot, cdf_logfit_plot, cdf_burrfit_plot])


for i in range(len(X)):
    plt.plot((X[i], X[i]), (fra[i], cdf_logfit[i]), ':', color="red", alpha=0.5)
    #Omitido para não poluir muito o gráfico
    plt.plot((X[i], X[i]), (fra[i], cdf_burrfit[i]), '--', color="blue", alpha=0.5)
    
```

Frequência relativa acumulada X  Função de distribuição acumulada.
Neste gráfico podemos perceber que as duas curvas de distribuição teóricas são bem semelhantes à dos dados porém a vermelha (da LogNormal) parece ser a que se distancia menos por maiores intervalos da curva.

```python
# Kolmogorov-Smirnov goodness-of-fit test
ks_logfit = max(abs(fra-cdf_logfit))
ks_burrfit = max(abs(fra-cdf_burrfit))


print("Teste K-S para a LogNormal com parâmetros da função fit:")
print(ks_logfit)
print("Teste K-S para a distribuiçã Burr com parâmetros da função fit:")
print(ks_burrfit)
```

A máxima discrepância entre a frequência relativa acumulada encontrada nos dados e a cumulativa da distribuição teórica é a da distribuição Burr, por isso o melhor ajuste é oferecido pela distribuição Logarítmica normal.
