### Itens 1 a 3 - Resolução:

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from scipy import stats

print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
Tabela = pd.read_csv('the-counted-2015.csv', encoding="windows-1252", sep= r'","', names = ["uid", "name", "age", "gender", "ethnicity", "month", "day", "year", "streetaddress", "city", "state", "classification", "law enforcement agency", "armed"])
tabela = Tabela.drop([0])
tabela
```

```python
Idade = tabela.age.convert_objects(convert_numeric=True)
Idade = Idade[(Idade > 0) & (Idade < 100)]
Idade.plot.hist()
```

Segundo o formato do histograma obtido com os dados do DataSet estudado, é razoável pensar que as curvas da distribuição "Burr" e da distribuição "Logarítmica normal" representam o histograma obtido nesse estudo.


### Gráfico das Distribuições Logarítmica e Burr com parâmetros estimados

```python
#Tentativa de estimar os valores dos parâmetros com LogNormal e Burr
plt.title('Estimativa dos valores')
arange = np.arange(5, 88, 1)
Idade.plot.hist(normed = 1, color = "gray")
TentaLog = stats.lognorm.pdf(arange, s = 0.37, loc = -1.9, scale = 35)
plt.plot(arange, TentaLog, "-", color = "purple", label = "LogNormal")
TentaB = stats.burr.pdf(arange, 3.81, 1, loc= -0.9, scale= 31)
plt.plot(arange, TentaB, "-", color = "orange", label="Burr")
plt.legend()
```

### Distribuição Logarítmica Normal com os valores da função fit()

```python
#Pegar dados com o fit
x = stats.lognorm.fit(Idade)
x
```

```python
#Distribuição com Log Normal
plt.title('Distribuição com Log Normal e .fit()')
Idade.plot.hist(normed = 1, color = "gray")
arange = np.arange(5, 88, 1)
Log = stats.lognorm.pdf(arange, scale = x[2], loc = x[1], s = x[0])
plt.plot(arange, Log, "-", color = "red")
plt.show()
```

### Distribuição Burr com os valores da função fit()

```python
# fit com Burr
z = stats.burr.fit(Idade)
z
```

```python
#Distribuição com Burr
plt.title('Distribuição com Burr e .fit()')
Idade.plot.hist(normed = 1, color="gray")
Burr = stats.burr.pdf(arange, z[0], z[1], loc= z[2], scale= z[3])
plt.plot(arange, Burr, "-", color = "blue")
plt.show()
```

### Comparação entre as duas distribuições com os parâmetros da função fit()

```python
plt.title('Comparação entre as duas distribuições com .fit()')
Idade.plot.hist(normed = 1, color = "gray")
arange = np.arange(5, 88, 1)
Log = stats.lognorm.pdf(arange, scale = x[2], loc = x[1], s = x[0])
plt.plot(arange, Log, "-", color = "red", label = "LogNormal")
Burr = stats.burr.pdf(arange, z[0], z[1], loc= z[2], scale= z[3])
plt.plot(arange, Burr, "-", color = "blue", label = "Burr")
plt.legend()
```

### Itens 10 e 11 - Resolução:

### Teste de aderência - frequências acumuladas

```python
# fra e FDA
cdf_logfit = stats.lognorm.cdf(X, scale = x[2], loc = x[1], s = x[0]) #FDA
cdf_burrfit = stats.burr.cdf(X, c = z[0], d = z[1], loc= z[2], scale= z[3])


fig = plt.figure(figsize=(20,12))

plt.title("Cumulativa: Dados x LogNormal x Burr")

dados_fra_plot, = plt.plot(X, fra, '-', color='black', label="F.R.A dos dados")
cdf_logfit_plot,   = plt.plot(X, cdf_logfit, 'o', color='red', label = "Cumulativa da LogNormal")
cdf_burrfit_plot, = plt.plot(X, cdf_burrfit, 'o', color='blue', label = "Cumulativa da Burr")

plt.legend(handles=[dados_fra_plot, cdf_logfit_plot, cdf_burrfit_plot], loc = 'upper left')


for i in range(len(X)):
    plt.plot((X[i], X[i]), (fra[i], cdf_logfit[i]), ':', color="red", alpha=0.2)
    plt.plot((X[i], X[i]), (fra[i], cdf_burrfit[i]), '--', color="blue", alpha=0.2)
    
```

Frequência relativa acumulada X  Função de distribuição acumulada.
Neste gráfico podemos perceber que as duas curvas de distribuição teóricas são bem semelhantes à dos dados porém a vermelha (da LogNormal) parece ser a que se distancia menos por maiores intervalos da curva.

### Teste de aderência de Kolmogorov-Smirnov

```python
# Kolmogorov-Smirnov goodness-of-fit test
ks_logfit = max(abs(fra-cdf_logfit))
ks_burrfit = max(abs(fra-cdf_burrfit))


print("Teste K-S para a LogNormal com parâmetros da função fit:")
print(ks_logfit)
print("Teste K-S para a distribuição Burr com parâmetros da função fit:")
print(ks_burrfit)
```

A máxima discrepância entre a frequência relativa acumulada encontrada nos dados e a cumulativa da distribuição teórica é a da distribuição Burr, por isso o melhor ajuste é oferecido pela distribuição Logarítmica normal.

### Erro quadrático nas CDFs

```python
# Erro quadrático nas CDFs
erro_quad_cdf_logfit = sum((fra - cdf_logfit)**2)
erro_quad_cdf_burrfit = sum((fra - cdf_burrfit)**2)
print("Erro quadrático da CDF da LogNormal X F.R.A:")
print(erro_quad_cdf_logfit)
print("Erro quadrático da CDF da Distribuição Burr X F.R.A:")
print(erro_quad_cdf_burrfit)
```

### Diferenças absolutas 

```python
# Diferenças absolutas
soma_difs_cdf_logfit = sum( abs(fra - cdf_logfit))
soma_difs_cdf_burrfit = sum( abs(fra - cdf_burrfit))
print("Erro da soma do módulo das diferenças da LogNormal versus F.R.A:")
print(soma_difs_cdf_logfit)
print("Erro da soma do módulo das diferenças da Distr. Burr versus F.R.A:")
print(soma_difs_cdf_burrfit)
```
